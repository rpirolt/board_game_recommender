```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Step 1: compute rating counts ---
game_counts = df_ratings['BGGId'].value_counts()
user_counts = df_ratings['Username'].value_counts()

# --- Step 2: define game rating cutoffs to test ---
cutoffs = [1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]

# --- Step 3: calculate % retained for each cutoff (only filtering on games) ---
results = []

total_games = len(game_counts)
total_users = len(user_counts)

for cut in cutoffs:
    # keep only games with >= cut ratings
    valid_games = game_counts[game_counts >= cut].index
    filtered = df_ratings[df_ratings['BGGId'].isin(valid_games)]
    
    # recompute which users still have at least one rating after removing low-rated games
    remaining_games = filtered['BGGId'].nunique()
    remaining_users = filtered['Username'].nunique()
    
    results.append({
        'min_game_ratings': cut,
        'percent_games_kept': remaining_games / total_games * 100,
        'percent_users_kept': remaining_users / total_users * 100
    })

# --- Step 4: convert to DataFrame ---
retention_df = pd.DataFrame(results)
display(retention_df)

# --- Step 5: plot both curves ---
plt.figure(figsize=(8,5))
plt.plot(retention_df['min_game_ratings'], retention_df['percent_games_kept'], marker='o', label='Games')
plt.plot(retention_df['min_game_ratings'], retention_df['percent_users_kept'], marker='o', color='orange', label='Users')

plt.xscale('log')
plt.xlabel('Minimum Ratings per Game (Game Filter Only)')
plt.ylabel('% of Unique Entities Retained')
plt.title('Retention Curves: Impact of Filtering Low-Rated Games')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Step 1: compute rating counts ---
user_counts = df_ratings['Username'].value_counts()
game_counts = df_ratings['BGGId'].value_counts()

# --- Step 2: define user rating cutoffs to test ---
cutoffs = [1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]

# --- Step 3: calculate % retained for each cutoff (only filtering on users) ---
results = []

total_users = len(user_counts)
total_games = len(game_counts)

for cut in cutoffs:
    # keep only users with >= cut ratings
    valid_users = user_counts[user_counts >= cut].index
    filtered = df_ratings[df_ratings['Username'].isin(valid_users)]
    
    # recompute how many users/games remain after removing low-activity users
    remaining_users = filtered['Username'].nunique()
    remaining_games = filtered['BGGId'].nunique()
    
    results.append({
        'min_user_ratings': cut,
        'percent_users_kept': remaining_users / total_users * 100,
        'percent_games_kept': remaining_games / total_games * 100
    })

# --- Step 4: convert to DataFrame ---
retention_df = pd.DataFrame(results)
display(retention_df)

# --- Step 5: plot both curves ---
plt.figure(figsize=(8,5))
plt.plot(retention_df['min_user_ratings'], retention_df['percent_users_kept'], marker='o', label='Users', color='orange')
plt.plot(retention_df['min_user_ratings'], retention_df['percent_games_kept'], marker='o', label='Games', color='steelblue')

plt.xscale('log')
plt.xlabel('Minimum Ratings per User (User Filter Only)')
plt.ylabel('% of Unique Entities Retained')
plt.title('Retention Curves: Impact of Filtering Low-Activity Users')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Since there's no perfect cutoff, I'm tentatively using
# 20 ratings per game and 2 ratings per user as thresholds. 

# 20 ratings per game only removes 1 game from the dataset.
# 2 ratings per user ensures CF always has at least one other 
# rating to work with.

# This seems like the lowest cutoff that enables CF while retaining most data.
```

```python
# --- Step 1: Define chosen thresholds ---
MIN_GAME_RATINGS = 20
MIN_USER_RATINGS = 5

# --- Step 2: Compute rating counts again (fresh, from full df_ratings) ---
game_counts = df_ratings['BGGId'].value_counts()
user_counts = df_ratings['Username'].value_counts()

# --- Step 3: Filter based on both conditions ---
valid_games = game_counts[game_counts >= MIN_GAME_RATINGS].index
valid_users = user_counts[user_counts >= MIN_USER_RATINGS].index

final_ratings_df = df_ratings[
    df_ratings['BGGId'].isin(valid_games) &
    df_ratings['Username'].isin(valid_users)
].copy()

print(f"Filter applied: min {MIN_GAME_RATINGS} ratings per game, min {MIN_USER_RATINGS} ratings per user")
# print(f"Original ratings: {len(df_ratings):,}")
# print(f"Filtered ratings: {len(final_ratings_df):,}")
# print(f"Retention: {len(final_ratings_df) / len(df_ratings) * 100:.2f}%")

# --- Step 4: Analyze final dataset ---
unique_games = final_ratings_df['BGGId'].nunique()
unique_users = final_ratings_df['Username'].nunique()
# print(f"\nUnique games: {unique_games:,}")
# print(f"Unique users: {unique_users:,}")

# --- Step 5: Visual sanity checks ---
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.hist(final_ratings_df['BGGId'].value_counts(), bins=50, color='steelblue', edgecolor='black')
plt.xlabel('Number of Ratings per Game')
plt.ylabel('Number of Games')
plt.title('Distribution of Ratings per Game (after filtering)')
plt.yscale('log')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,5))
plt.hist(final_ratings_df['Username'].value_counts(), bins=50, color='orange', edgecolor='black')
plt.xlabel('Number of Ratings per User')
plt.ylabel('Number of Users')
plt.title('Distribution of Ratings per User (after filtering)')
plt.yscale('log')
plt.tight_layout()
plt.show()

# --- Step 6: Optional summary table ---
summary = {
    'Metric': ['Total Ratings', 'Unique Games', 'Unique Users'],
    'Before Filter': [len(df_ratings), df_ratings["BGGId"].nunique(), df_ratings["Username"].nunique()],
    'After Filter': [len(final_ratings_df), unique_games, unique_users],
    'Retention (%)': [len(final_ratings_df) / len(df_ratings) * 100,
                      unique_games / df_ratings["BGGId"].nunique() * 100,
                      unique_users / df_ratings["Username"].nunique() * 100]
}
summary_df = pd.DataFrame(summary)
display(summary_df)

#print BGGIds of removed games
all_games = set(df_ratings['BGGId'].unique())
kept_games = set(final_ratings_df['BGGId'].unique())
removed_games = all_games - kept_games
removed_games_clean = [int(x) for x in removed_games]
print("Removed BGGIds:", removed_games_clean)
```

```python
# Next thing to look at is CBF and LLM features.

## LLM
# df_games Description

# Identify rows with missing or blank descriptions
missing_mask = df_games['Description'].isna() | (df_games['Description'].str.strip() == '')

# Count them
missing_descriptions = missing_mask.sum()
print(f"Number of games with missing descriptions: {missing_descriptions:,}")

# Show the corresponding BGGIds (and optionally names)
missing_games = df_games.loc[missing_mask, ['BGGId', 'Name']].reset_index(drop=True)
print("\nGames with missing descriptions:")
display(missing_games)
```

```python
## CBF
# df_data mechanics
# df_data boardgamecategory
# df_data gametype
# df_games GameWeight
# df_data best_numplayers
# df_data playingtime

# print number of missing values for each feature
features = [
    'mechanics',
    'boardgamecategory',
    'gametype',
    'GameWeight',
    'best_numplayers',
    'playingtime'
]
for feature in features:
    if feature in df_data.columns:
        missing_count = df_data[feature].isna().sum()
        print(f"Missing '{feature}' values in df_data: {missing_count:,}")
    elif feature in df_games.columns:
        missing_count = df_games[feature].isna().sum()
        print(f"Missing '{feature}' values in df_games: {missing_count:,}")
    else:
        print(f"Feature '{feature}' not found in either DataFrame.")

#get list of BGGids with any missing CBF features
removed_games = set()
for feature in features:
    if feature in df_data.columns:
        missing_bggids = df_data.loc[df_data[feature].isna(), 'id'].unique()
        removed_games.update(missing_bggids)
    elif feature in df_games.columns:
        missing_bggids = df_games.loc[df_games[feature].isna(), 'BGGId'].unique()
        removed_games.update(missing_bggids)
print(f"\nNumber of games with any missing CBF features: {len(removed_games):,}")
print(f"\nNumber of games with no missing CBF features: {df_games.shape[0] - len(removed_games):,}")
```

```python
### SUMMARY
# Limiting the user_ratings dataset to only include games with at least 20 
# ratings and users with at least 2 ratings results in a loss of only 1 game,
# retains 80% of users, and 99.6% of ratings. 

# Limiting the games dataset to only include games with descriptions for 
# the LLM model results in the removal of only 1 game.

# Limiting the CBF features to only include games with no missing values
# results in the removal of 9,905 games. gametype is the worst offender
# with 8,781 missing values. We can preserve most of the boardgame dataset
# by dropping the gametype feature from bgg_data in CBF.
```
